Do some machine learning on data found here http://mlr.cs.umass.edu/ml/machine-learning-databases/adult/using aimed at predicting whether or not a person will make over 50k a year. Do it in an IPython Notebook. Look in adult.data for the data, save it as a csv file. The adult.names file has information about the set. 

You may use Pandas if you want, google for a tutorial (there are many, and it's easy to learn). Or you can just use Numpy, or a combination of both.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('adult.data', header=0, na_values=[' ?'])
df.columns = [
    'age',
    'workclass',
    'fnlwgt',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'capital-gain',
    'capital-loss',
    'hours-per-week',
    'native-country',
    '50k?'
]

df, y = np.split(df, [14], axis=1)
```

2 pts Replace all empty numeric fields with the average of its column in this new column.

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

2 pts Convert non-numeric columns to columns with an integer representations (like we did with the decision trees. 

Note: for "?", please find the value that is most common in the column it is found and replace the "?" with it.

```python
df.fillna(df.mode().iloc[0], inplace=True)
print(df[df.isnull().any(axis=1)].shape)

# One-hot encoding doesn't play nice with some of the classifiers in terms of performance and accuracy
to_factorize = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
for column_name in to_factorize:
    df[column_name] = pd.factorize(df[column_name])[0]

y['50k?'] = pd.factorize(y['50k?'])[0]
```

2 pts Convert the data to a numpy array if it isn't one already.

```python
X = df.values
y = y.values.ravel()
```

2 pts Divide the data into a training and test set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

2 pts Run both a random forest and gaussian naive bayes on the training set.


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

clf_forest = RandomForestClassifier(n_estimators=8)
clf_forest.fit(X_train, y_train)
predicted_forest = clf_forest.predict(X_test)

clf_gnb = GaussianNB()
clf_gnb.fit(X_train, y_train)
predicted_gnb = clf_gnb.predict(X_test)
```

2 pts Compute the accuracy of predictions on the test set.

```python
from sklearn import metrics
print('Random Forest')
print('Accuracy on training data:', clf_forest.score(X_train, y_train))
print(metrics.classification_report(y_test, predicted_forest,
    target_names=['<50k', '>=50k']))

print('Gaussian Naive Bayes')
print('Accuracy on training data:', clf_gnb.score(X_train, y_train))
print(metrics.classification_report(y_test, predicted_gnb,
    target_names=['<50k', '>=50k']))
```

2 pts each extra credit: Use another additional classifier.

Used 18 classifiers

```python
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier, NearestCentroid, RadiusNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier, AdaBoostClassifier, BaggingClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.dummy import DummyClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import PassiveAggressiveClassifier, SGDClassifier
from sklearn.naive_bayes import BernoulliNB, MultinomialNB

classifiers = [
    KNeighborsClassifier(3),
    NearestCentroid(),
    RadiusNeighborsClassifier(radius=1000, outlier_label=0),
    LinearSVC(dual=False),
    DecisionTreeClassifier(min_samples_split=1000),
    ExtraTreesClassifier(n_estimators=10, max_depth=None, min_samples_split=100),
    MLPClassifier(alpha=0.1, hidden_layer_sizes=(10,)),
    AdaBoostClassifier(),
    BaggingClassifier(),
    GradientBoostingClassifier(),
    DummyClassifier(),
    LinearDiscriminantAnalysis(solver='lsqr'),
    QuadraticDiscriminantAnalysis(),
    PassiveAggressiveClassifier(max_iter=1000, tol=None),
    SGDClassifier(max_iter=1000, tol=1e-3),
    BernoulliNB(),
    MultinomialNB()
]

voting = VotingClassifier(estimators=[(classifier.__class__.__name__, classifier) for classifier in classifiers])
classifiers.append(voting)

print('Running', len(classifiers), 'classifiers\n')

for clf in classifiers:
    clf.fit(X_train, y_train)
    predicted = clf.predict(X_test)
    print(clf)
    print('Accuracy on training data:', clf.score(X_train, y_train))
    print(metrics.classification_report(y_test, predicted,
        target_names=['<50k', '>=50k']))
```
